## Test 1: Pearson correlation coefficeint

Description of Test and Assumptions:

Measures the strength and direction of a LINEAR correlation between two variables

- Assumes a random sample
- Both variables are continuous 
- Data contains paired examples(each subject has x and y values such as weight and height)
- There is independence between subjects(i.e. subjects are not related)
- Both x and y values are assumed to be on a normal distribution
- There are no outliers in the data
- No correlation if r ~ or < probable error
- Correlation coefficient is ‘certain’ if r is at least 6 times greater than probable error





**The output of this test is the correlation coefficient "r"**

https://www.youtube.com/watch?v=e4ApDqG6MGE&t=5s

**SciPy Documentation Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html** 

```python
r = stats.pearsonr(x, y)
n = len(y)
R = r[0]*r[0]
Error = .674 *((1-R)/(np.sqrt(n)))
print("The correlation coefficient is: " + str(r[0]) + " +/- " + str(Error))
print("The calculated p value is " + str(r[1]))
```
